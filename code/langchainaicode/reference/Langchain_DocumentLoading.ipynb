```python
from langchain.document_loaders import OnlinePDFLoader

loader = OnlinePDFLoader("https://arxiv.org/pdf/1706.03762.pdf")
data = loader.load()
print(data)
```

### Loading & Splitting

As you can see above, we retrieve the whole document if we just call `load()`, if we want to split this document up so that is more appropriate for a LLM we can call `load_and_split()` and Langchain will split it for us.

We will discuss more advanced ways to split documents in the next lecture on document splitting.

```python
data = loader.load_and_split()
print(data)
```

```python
len(data)
```

```python
data[0]
```

# CSV Loader


Langchain offers a CSV Loader to convert your data per row into a string (text). This is beneficial if you want to use natural language to query your CSV data.

```python
from langchain.document_loaders.csv_loader import CSVLoader


loader = CSVLoader(file_path='./reviews_data.csv')
data = loader.load()
data
```

# Loading Data From Any URL

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.udemy.com/courses/search/?src=ukw&q=langchain")
docs = loader.load()
docs
```

```python
import requests


resp = requests.get("https://www.udemy.com/courses/search/?src=ukw&q=langchain")
print(resp.text)
```
