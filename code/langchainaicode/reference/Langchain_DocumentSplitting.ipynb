# Character Splitting

Langchain has two main text splitters for general text:

- CharacterTextSplitter
- RecursiveCharacterTextSplitter

However, they suggest to use the Recursive text splitter for general text.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter, CharacterTextSplitter

chunk_size = 2000
chunk_overlap = 200

character_splitter = CharacterTextSplitter(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap
)

recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap,
    separators=["\n\n", "\n", " ", ""]
)
```

# Code Splitter

```python
from langchain.text_splitter import (
    RecursiveCharacterTextSplitter,
    Language,
)

# You can also see the separators used for a given language
RecursiveCharacterTextSplitter.get_separators_for_language(Language.PYTHON)
```

```python
PYTHON_CODE = """
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Splitting the array
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursive sorting
    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    # Merging the sorted arrays
    return merge(left_sorted, right_sorted)

def merge(left, right):
    merged = []
    left_index, right_index = 0, 0

    # Compare each index of the two halves and append the smaller element
    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    # If left half still has elements, append them
    while left_index < len(left):
        merged.append(left[left_index])
        left_index += 1

    # If right half still has elements, append them
    while right_index < len(right):
        merged.append(right[right_index])
        right_index += 1

    return merged

# Test
arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = merge_sort(arr)
print(sorted_arr)  # Expected output: [3, 9, 10, 27, 38, 43, 82]
"""
python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON, chunk_size=500, chunk_overlap=0
)
python_docs = python_splitter.create_documents([PYTHON_CODE])
python_docs
```
